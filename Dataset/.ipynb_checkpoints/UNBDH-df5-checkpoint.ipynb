```python
def cat_num (df):
    for typ,col in zip(df.dtypes.values,df.dtypes.index):
        if typ == "object":
             cat_col.append(col)
        else:
            num_col.append(col);

def cat_count(df):
    cat_columns = df[cat_col]
    for col in cat_columns:
        print(f'value counts dari column {col}')
        print(df[col].value_counts())
        print('---'*10, '\n');
```

```python
plt.figure(figsize=(15,25))
for i in range(0, len(num_col)):
    plt.subplot(14, 4, i+1)
    sns.boxplot(x=df5[num_col[i]], color='royalblue')
    plt.xlabel(num_col[i])
    plt.tight_layout()
```

```python
plt.figure(figsize= (15, 25))
for i in range(len(num_col)):
    plt.subplot(14, 4, i+1)
    sns.kdeplot(x = df5[num_col[i]])
    plt.tight_layout()
```

```python
var_corr = df5[num_col].corr()
fig = plt.figure(figsize = (20,18))
sns.heatmap(var_corr, cmap='plasma', annot=True, mask = (abs(var_corr) <= 0.2))
```

Observation:
- Both renewable and nonrenewable energy consumption tends to negatively correlate with Poverty, Hunger, and Inequalities
- The vicious cycle, including poverty, hunger, and inequalities tends to have a positive correlation with each other
- The rise of diesel fuel is negatively correlated with the vicious cycle and positively correlated with the rise of some renewable energy consumption.
